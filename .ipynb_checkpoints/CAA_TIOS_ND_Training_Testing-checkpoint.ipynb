# CAA-TIOS-ND Training and Testing Notebook

This notebook trains and tests the Constraint-Aware Adaptive Travel Itinerary Optimization System with Intelligent Nearby Discovery.

It uses `train_caa_tios_nd.py` as the reusable engine and separates the workflow into setup, training, scoring, and itinerary prediction.

## 1. Setup

```python
from pathlib import Path
from types import SimpleNamespace
import json
import joblib
import numpy as np
import pandas as pd

import train_caa_tios_nd as engine

CSV_PATH = Path("/content/Cleaned_Travel_Dataset.csv")
MODEL_PATH = Path("caa_tios_nd_model.joblib")
REPORT_PATH = Path("training_report.json")

assert CSV_PATH.exists(), f"CSV not found: {CSV_PATH}"
print("Dataset:", CSV_PATH)
print("Model output:", MODEL_PATH.resolve())
```

## 2. Load Dataset and Vocabulary

```python
df = engine.read_dataset(CSV_PATH)
cities, locations = engine.build_vocabularies(df)
examples = engine.build_stop_level_examples(df, locations)

print("Raw dataset shape:", df.shape)
print("Stop-level training examples:", examples.shape)
print("Cities:", cities)
print("Location count:", len(locations))
pd.DataFrame({"locations": locations}).head(20)
```

## 3. Training Code

This cell trains the full artifact and saves both the model and training report.

```python
train_args = SimpleNamespace(
    csv=str(CSV_PATH),
    model_out=str(MODEL_PATH),
    report_out=str(REPORT_PATH),
)

engine.train(train_args)

print("Saved model:", MODEL_PATH.resolve())
print("Saved report:", REPORT_PATH.resolve())
```

## 4. Training Report Metrics

The training function already performs an internal train/test split and records metrics for route ranking, timing, status, and warning prediction.

```python
report = json.loads(REPORT_PATH.read_text(encoding="utf-8"))
print(json.dumps(report["metrics"], indent=2))

metric_rows = []
for model_name, values in report["metrics"].items():
    row = {"model": model_name, **values}
    if "classes" in row:
        row["classes"] = ", ".join(map(str, row["classes"]))
    metric_rows.append(row)

pd.DataFrame(metric_rows)
```

## 5. Detailed Test Scores

This section recomputes holdout metrics with extra scores: MAE, RMSE, R2 for regression tasks and accuracy, precision, recall, F1, confusion matrix, and classification report for classification tasks.

```python
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    f1_score,
    mean_absolute_error,
    mean_squared_error,
    precision_score,
    r2_score,
    recall_score,
)
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

categorical_features = ["city", "priority", "location_name", "city_location"]
numeric_features = [
    "days",
    "slot_index",
    "lat",
    "lng",
    "rating",
    "open_min",
    "close_min",
    "visit_duration",
    "mandatory",
    "window_span",
    "window_slack",
    "location_count",
    "mandatory_count",
    "avg_rating",
    "max_rating",
    "total_visit_duration",
    "earliest_open_min",
    "latest_close_min",
    "geo_span_km",
]
feature_cols = categorical_features + numeric_features

def regression_scores(target_col):
    usable = examples[examples[target_col].notna()].copy()
    train_df, test_df = train_test_split(usable, test_size=0.2, random_state=42)
    model = Pipeline([
        ("preprocess", engine.build_preprocessor(categorical_features, numeric_features)),
        ("model", RandomForestRegressor(n_estimators=300, min_samples_leaf=2, random_state=42, n_jobs=-1)),
    ])
    model.fit(train_df[feature_cols], train_df[target_col].astype(float))
    pred = model.predict(test_df[feature_cols])
    y_true = test_df[target_col].astype(float)
    return {
        "target": target_col,
        "train_rows": len(train_df),
        "test_rows": len(test_df),
        "mae": mean_absolute_error(y_true, pred),
        "rmse": float(np.sqrt(mean_squared_error(y_true, pred))),
        "r2": r2_score(y_true, pred),
    }

def classification_scores(target_col):
    usable = examples[examples[target_col].notna()].copy()
    usable = usable[usable[target_col].astype(str) != ""]
    train_df, test_df = train_test_split(usable, test_size=0.2, random_state=42)
    classes = sorted(set(train_df[target_col].astype(str)))
    if len(classes) < 2:
        estimator = DummyClassifier(strategy="most_frequent")
    else:
        estimator = RandomForestClassifier(
            n_estimators=300,
            min_samples_leaf=2,
            random_state=42,
            class_weight="balanced_subsample",
            n_jobs=-1,
        )
    model = Pipeline([
        ("preprocess", engine.build_preprocessor(categorical_features, numeric_features)),
        ("model", estimator),
    ])
    model.fit(train_df[feature_cols], train_df[target_col].astype(str))
    pred = model.predict(test_df[feature_cols])
    y_true = test_df[target_col].astype(str)
    labels = sorted(set(y_true) | set(map(str, pred)))
    return {
        "target": target_col,
        "train_rows": len(train_df),
        "test_rows": len(test_df),
        "classes": labels,
        "accuracy": accuracy_score(y_true, pred),
        "precision_macro": precision_score(y_true, pred, average="macro", zero_division=0),
        "recall_macro": recall_score(y_true, pred, average="macro", zero_division=0),
        "f1_macro": f1_score(y_true, pred, average="macro", zero_division=0),
        "f1_weighted": f1_score(y_true, pred, average="weighted", zero_division=0),
        "confusion_matrix": confusion_matrix(y_true, pred, labels=labels).tolist(),
        "classification_report": classification_report(y_true, pred, labels=labels, zero_division=0, output_dict=True),
    }

detailed_scores = [
    regression_scores("target_position"),
    regression_scores("target_arrival_min"),
    regression_scores("target_departure_min"),
    classification_scores("target_status"),
    classification_scores("target_location_warning"),
]

summary_rows = []
for item in detailed_scores:
    summary_rows.append({k: v for k, v in item.items() if k not in {"confusion_matrix", "classification_report"}})

pd.DataFrame(summary_rows)
```

```python
for item in detailed_scores:
    if "confusion_matrix" in item:
        print("\nTarget:", item["target"])
        print("Classes:", item["classes"])
        print("Confusion matrix:")
        print(np.array(item["confusion_matrix"]))
```
